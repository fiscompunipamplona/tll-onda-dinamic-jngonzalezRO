```python
from pylab import imshow,show,xlabel,ylabel,plot,xlim,ylim,hot,figure,jet,flag,plasma,hsv,inferno,viridis,cool
import numpy as np
from IPython import display as disp
import matplotlib.pyplot as plt
from numpy import linspace,sin,cos,sqrt,loadtxt
salida1=open("ondas_esta3.dat","w")
u=150# numero de espacios
v=30 #
t=100
#fuentes de onda
x=20;y=15
x2=10;y2=15
x1=linspace(0,v,u)
y1=linspace(0,v,u)
X,Y=np.meshgrid(x1,y1)
info=np.zeros((u,u,t),dtype=float)
xlim(0.,v)
ylim(0.,v)
G1=1
No=5
k=(2*3.1416)/No

w=100
for t in range(0,t):   
    r1 = (((X-x)**2)+((Y-y)**2))**0.5
    #print(G)
    G = G1*sin(k*r1+w*t)
    r2 =(((X-x2)**2)+((Y-y2)**2))**0.5
    G2 = G1*sin(k*r2+w*t)
    G3  = G2+G
    plt.gca().cla()#
    info[:,:,t]=G3
    imshow(G3,origin="lower")
    inferno()
    disp.clear_output(wait=True)#
    disp.display(plt.gcf())#
    #time.sleep(0.1)
```

```python
for r in range(t):
    salida1.write('#Matriz %d \n' %(r))
    for i in range(u):
        for j in range(u):
            salida1.write('%0.4f ' %(info[i,j,r]))
        salida1.write('\n')
salida1.close()               
```
